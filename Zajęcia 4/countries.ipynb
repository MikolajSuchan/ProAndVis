```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('european-countries.csv')
countries=pd.DataFrame(data)
countries
```

```python
chart=countries.query('Region=="Central Europe"').sort_values('Population')
chart
```

```python
chart.index=['Slovenia','Slovakia','Austria','Hungary','Chech Republic','Poland']
chart['Population'].plot(kind='bar')
```

```python
simple_data=pd.DataFrame(data)
chart2=simple_data.groupby('Region')['Area'].sum()
chart2.plot(kind='pie',autopct='%1.0f%%')
```
