```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('employees.csv')
chart=pd.DataFrame(data)
chart
```

```python
chart.groupby('sex')['age'].mean().plot(kind='bar',title='Average age based on sex',xlabel='Sex',ylabel='Average age')
```

```python
chart.plot(subplots=True,kind='hist')
```
